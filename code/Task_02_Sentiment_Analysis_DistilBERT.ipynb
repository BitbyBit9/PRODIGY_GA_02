```python
from transformers import pipeline

sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

print("Sentiment analysis model loaded successfully.")
```

```python
text = "I really enjoyed learning about transformer models."
result = sentiment_analyzer(text)

print(result)
```

```python
sentences = [
    "I absolutely loved this internship experience.",
    "The product quality was very poor and disappointing.",
    "Learning transformers was challenging but rewarding.",
    "I would not recommend this service to anyone."
]

results = sentiment_analyzer(sentences)

for sentence, result in zip(sentences, results):
    print(f"Text: {sentence}")
    print(f"Sentiment: {result['label']}, Confidence: {result['score']:.4f}")
    print("-" * 60)
```

```python
with open("sentiment_output.txt", "w") as file:
    for sentence, result in zip(sentences, results):
        file.write(f"Text: {sentence}\n")
        file.write(f"Sentiment: {result['label']}, Confidence: {result['score']:.4f}\n")
        file.write("-" * 50 + "\n")

print("Output saved successfully.")
```
